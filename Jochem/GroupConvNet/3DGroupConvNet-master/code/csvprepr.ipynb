```python
import warnings; warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
import os.path as osp
import os
import tensorflow as tf
```

```python
df = pd.read_csv('/data_local/deeplearning/ukbb_LC/pheno/Lifetime_Depression.csv')
ids = df['ID']
df['Group']
```

```python
files = '/data_local/deeplearning/ukbb/T1_MNI_2mm'
fs = []
labels = []
index = 0
ignored = 0
new_ids = []
fs_ignored = []
for i in ids:
    f = osp.join(files,'{}_T1_MNI_2mm.nii.gz'.format(i))
    if osp.isfile(f):
        fs.append(f)
        if df['Group'].iloc[index] == 'Control':
            s = 0
        else:
            s = 1
        labels.append(s)
        new_ids.append(i)
    else:
        print (f)
        fs_ignored.append(i)
        ignored +=1
    index +=1
    


    
    
```

```python
print(len(fs))
print(len(labels))
print(len(new_ids))
```

```python
dic = {'ids':new_ids,'paths': fs,'label': labels}
df_new = pd.DataFrame(dic)
df_new.to_csv('csvfiles/LifetimeDepression/Lifetime_Depression.csv')
```

```python
train, validate, test = np.split(df_new.sample(frac=1), [int(.7*len(df_new)), int(.8*len(df_new))])
train.to_csv('csvfiles/LifetimeDepression/train.csv')
validate.to_csv('csvfiles/LifetimeDepression/validate.csv')
test.to_csv('csvfiles/LifetimeDepression/test.csv')
```
