```python
import warnings; warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
import os.path as osp
import os
import tensorflow as tf
```

```python
path = '/data_local/deeplearning/UKBiobank/Release_17DEC2018/metadata.csv'
df = pd.read_csv(path)
ids = df['f.eid']
sex = df['f.31.0.0']
```

```python
files = '/data_local/deeplearning/ukbb/T1_MNI_2mm'
fs = []
sexs = []
index = 0
ignored = 0
new_ids = []
for i in ids:
    f = osp.join(files,'{}_T1_MNI_2mm.npy'.format(i))
    if osp.isfile(f):
        fs.append(f)
        if df['f.31.0.0'].iloc[index] == 'Male':
            s = 0
        else:
            s = 1
        sexs.append(s)
        new_ids.append(i)
    else:
        ignored +=1
    index +=1
    
print (ignored)

    
    
```

```python
print(len(fs))
print(len(sexs))
print(len(new_ids))
```

```python
dic = {'ids':new_ids,'paths': fs,'sex': sexs}
df = pd.DataFrame(dic)
df.to_csv('Encoded_Gender_Classification.csv')
```
